```python
import xml.etree.ElementTree as ET
from pathlib import Path
```

```python
##running
PATH = Path("D:/Datasets/Brainwaves/ML/data/")
import re
termination_currency1 = {"EUR" : [],"USD" : [], "GBP" : [],"AUD" : []}
currency = r'[\"\']Termination Currency[\"\']'
euro_files = []
for xmlfile in PATH.iterdir():
    if str(xmlfile).split(".")[-1] == "xml":
        tree = ET.parse(xmlfile)
        xmlroot = tree.getroot()
        _start = 0
        temp_str = ""
        for child in xmlroot.iter(tag=xmlroot.tag.split("}")[0]+"}"+"formatting"):
            if  re.search(currency ,child.text) :
                _start = 1
            
            if _start != 0 and _start < 20 :
                temp_str += child.text
                _start+=1
        if "Euro" in temp_str or "EUR" in temp_str:
            termination_currency1["EUR"].append(str(xmlfile).split("\\")[-1].split(".")[0])
            
        if "Australia" in temp_str or "australia" in temp_str:
            termination_currency1["AUD"].append(str(xmlfile).split("\\")[-1].split(".")[0])
            
        else : 
            termination_currency1["USD"].append(str(xmlfile).split("\\")[-1].split(".")[0])

answer=pd.DataFrame()
for key in termination_currency1.keys():
    temp=pd.DataFrame()
    temp['File_Name']=np.array(termination_currency1[key])
    temp['TC_currency']=key
    answer=pd.concat([answer,temp])
    
answer             
```

```python
PATH = Path("D:/Datasets/Brainwaves/ML/public_test_data/")
import re
termination_currency1 = {"EUR" : [],"USD" : [], "GBP" : [],"AUD" : []}
currency = r'[\"\']Termination Currency[\"\']'
euro_files = []
for xmlfile in PATH.iterdir():
    if str(xmlfile).split(".")[-1] == "xml":
        tree = ET.parse(xmlfile)
        xmlroot = tree.getroot()
        _start = 0
        temp_str = ""
        for child in xmlroot.iter(tag=xmlroot.tag.split("}")[0]+"}"+"formatting"):
            if  re.search(currency ,child.text) :
                _start = 1
            
            if _start != 0 and _start < 20 :
                temp_str += child.text
                _start+=1

        if "Euro" in temp_str or "EUR" in temp_str:
            termination_currency1["EUR"].append(str(xmlfile).split("\\")[-1].split(".")[0])
            
        elif "Australia" in temp_str or "australia" in temp_str:
            termination_currency1["AUD"].append(str(xmlfile).split("\\")[-1].split(".")[0])
            
        else : 
            termination_currency1["USD"].append(str(xmlfile).split("\\")[-1].split(".")[0])
             
test_answer=pd.DataFrame()
for key in termination_currency1.keys():
    temp=pd.DataFrame()
    temp['File_Name']=np.array(termination_currency1[key])
    temp['TC_currency']=key
    test_answer=pd.concat([test_answer,temp])
    
test_answer                           
```

```python
def extract_government_law(loc):
    #all_laws = ["new york", "england", "south wales", "english law", "england wales","hong kong","NotFound"]
    all_laws = ["new york", "england", "south wales", "english law", "england wales","hong kong","NotFound"]
    mapping_dict = {"new york" : "Laws of the state of New York","england": "Laws of England","south wales" : "State of New South Wales","english law":"English Law","england wales" : "Laws of England and Wales", "hong kong" : "Laws of the Hong Kong Special Administrative region of the People's Republic of China ","NotFound": "NotFound"}

    tree=ET.parse(loc)
    root=tree.getroot()
    base=root[0].tag.split("}")[0]
    c=0
    z=0
    roots=[]
    data=[]
    content = ""
    for child in root.iter(base+"}"+'par'):
        rt=child
        for elem in rt.iter(base+"}"+"formatting"):
            if "governing law" in elem.text.lower():
                roots.append(rt)

    for parent in roots:
        for elem in parent.iter(base+"}"+"formatting"):
             content += elem.text.lower()
                
    
    if "england" in content and "wales" in content :
        return mapping_dict["england wales"]
      
    #print(loc," : ",content,"\n")
    for law in all_laws:
        if law in content :
            return mapping_dict[law]
    
    return mapping_dict["NotFound"]
```

```python
law_ans = []
for name in train_labels["File_Name"]:
    law_ans.append(extract_government_law("D:/Datasets/Brainwaves/ML/data/"+name+".xml"))
    
    
law_ans_test = []
for name in test_answer["File_Name"]:
    law_ans_test.append(extract_government_law("D:/Datasets/Brainwaves/ML/public_test_data/"+name+".xml"))
```
